```python
#7. Test the Analysis of Variance between etest_p and mba_p at signifance level 5%.(Make decision using Hypothesis Testing)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('etest_p ~ mba_p', data=dataset).fit()

anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
#Answer:
#since the ANOVA p-value is 0.001293, which is less than 0.05, the relationship is statistically significant.
#Therefore, we reject H₀ and conclude that mba_p and etest_p are significantly related.
```

```python
#8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) 
#with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing)
```

```python
from scipy.stats import ttest_ind

group1 = dataset[dataset["degree_t"] == "Sci&Tech"]["salary"]
group2 = dataset[dataset["specialisation"] == "Mkt&HR"]["salary"]

ttest_ind(group1, group2)
```

```python
from scipy.stats import ttest_ind

group1 = dataset[(dataset["degree_t"] == "Sci&Tech") &(dataset["salary"] > 0)]["salary"]

group2 = dataset[(dataset["specialisation"] == "Mkt&HR") & (dataset["salary"] > 0)]["salary"]

result = ttest_ind(group1, group2)

print(result)
```

```python
#Answer Since the p-value (0.007) is less than 0.05, we reject the null hypothesis. 
#Therefore, there is a statistically significant difference in salary between students in Sci&Tech and students in Mkt&HR
```

```python
#10)What is the probability Density Function of the salary range from 700000 to 900000?
```

```python
from scipy.stats import norm

mean = dataset["salary"].mean()
std = dataset["salary"].std()

probability = norm.cdf(900000, mean, std) - norm.cdf(700000, mean, std)

print("Probability:", probability)
print("Percentage:", probability * 100)
```

```python
def get_probability(dataset, startrange, endrange):
    from scipy.stats import norm
    import seaborn as sns
    from matplotlib import pyplot

    sns.distplot(dataset, kde=True,kde_kws={'color':'blue'},color='green')

    pyplot.axvline(startrange, color='red')
    pyplot.axvline(endrange, color='red')

    mean = dataset.mean()
    std = dataset.std()

    print("Mean =", mean)
    print("Std =", std)

    prob = norm.cdf(endrange, mean, std) - norm.cdf(startrange, mean, std)

    print("Probability:", prob)
    return prob
```

```python
get_probability(dataset["salary"], 700000, 900000)
```

```python
#11)Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.
#(Make decision using Hypothesis Testing)
```

```python
from scipy.stats import ttest_rel

SciTech = dataset[dataset["degree_t"]=="Sci&Tech"]["etest_p"]
SciTech1 = dataset[dataset["degree_t"]=="Sci&Tech"]["mba_p"]

stat,pvalue=ttest_rel(SciTech, SciTech1)
print("P-value:", pvalue)
if pvalue < 0.05:
    print("Reject H0")
    print("There is a significant difference between etest_p and mba_p")
else:
    print("Fail to Reject H0")
    print("There is no significant difference between etest_p and mba_p")
```
